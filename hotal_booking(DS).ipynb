```python
# Introduction

This project analyzes hotel booking data to understand customer behavior, booking patterns, and cancellation trends. 
The dataset contains various features such as booking details, stay duration, and customer information, including some missing values.

The objective of this project is to clean the data, perform analysis using graphs, and create new features for better understanding. 
This project demonstrates basic data science steps including data preprocessing, visualization, and exploratory data analysis.
```

```python
# Data Cleaning
```

```python
import pandas as pd

#Load Dataset
df = pd.read_csv("hotel_bookings.csv")
df
```

```python
# Display the top 5 rows of the dataset
df.head()
```

```python
# Display the last 5 rows of the dataset
df.tail()
```

```python
# Check duplicate Columns
df.columns.duplicated()
```

```python
# Check duplicate Rows
df.duplicated().sum()
```

```python
# Checking and removing duplicate rows to maintain data integrity
df = df.drop_duplicates()
```

```python
# View dataset information
df.info()
```

```python
# Check null values 
df.isnull().sum()
```

```python
# Drop rows that contain null (missing) values
df.drop(['company', 'agent'], axis=1, inplace=True)
```

```python
# Check null values 
df.isnull().sum()
```

```python
# Filling Missing Values
df['children'].fillna(0, inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
# Check null values 
df.isnull().sum()
```

```python
print("Missing values handled successfully")
```

```python
# This shows statistical summary of dataset
df.describe()
```

```python
# Visualization
```

```python
import matplotlib.pyplot as plt

# Most bookings are in resort/city hotel by using Bar Graph
df['hotel'].value_counts().plot(kind='bar')
plt.title("Hotel Type Distribution")
plt.show()
```

```python
# We analyzed cancellation behavior by using Histogram
df['is_canceled'].value_counts().plot(kind='bar')
plt.title("Booking Cancellation")
plt.show()
```

```python
# Shows booking time pattern
plt.hist(df['lead_time'])
plt.title("Lead Time Distribution")
plt.show()
```

```python
- Most bookings are from city hotels
- Cancellation rate is significant
- Customers book in advance (lead time analysis)
```

```python
# Most bookings come from these countries (Top Countries Analysis)
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

```python
# Majority customers are transient (Customer Type Analysis)
df['customer_type'].value_counts().plot(kind='bar')
plt.title("Customer Type Distribution")
plt.show()
```

```python
# This plot shows average hotel price for each month
avg_price = df.groupby('arrival_date_month')['adr'].mean()

avg_price.plot()
plt.xlabel("Month")
plt.ylabel("Average Price (ADR)")
plt.title("Average Price per Month")
plt.show()
```
